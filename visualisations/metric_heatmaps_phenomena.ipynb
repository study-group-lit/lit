```python
import json
import numpy as np

import sys
sys.path.append('..')
from semplotlib import plot_metric_heatmap_phenomena

metrics = ["f1", "accuracy", "matthews_correlation"]
phenomena = ["Synonym", "Antonym", "Hypernym", "Hyponym", "Co-Hyponym", "Quantifiers", "Numerals"]
```

## Important words

### Base & Filtered

```python
path_mapping = {
    "Base": "../scripts/metrics_by_phenomena/results/roberta-base-finetuned-mnli.json",
    "Filtered 3/3 longer": "../scripts/metrics_by_phenomena/results/roberta-base-finetuned-mnli-maximum_two_6.json"
}
models = ["Base", "Filtered 3/3 longer"]

for metric in metrics:
    metric_data = []
    for model in path_mapping.items():
        model_data = []
        with open(model[1], "r") as f:
            metrics_data = json.load(f)
            for phenomenon_data in metrics_data:
                model_data.append(phenomenon_data[metric])
        metric_data.append(model_data)
    print(metric)
    plot_metric_heatmap_phenomena(np.array(metric_data).T, models, phenomena, save_path=f"../report/images/metric_heatmaps_phenomena/important_words/base_filtered_{metric}.pdf")
```

### Hypothesis-Only

```python
path_mapping = {
    "Hypothesis-Only": "../scripts/metrics_by_phenomena_mnli_hypothesis_only/results/roberta-base-finetuned-mnli-hypothesis-only.json"
}
models = ["Hypothesis-Only"]

for metric in metrics:
    metric_data = []
    for model in path_mapping.items():
        model_data = []
        with open(model[1], "r") as f:
            metrics_data = json.load(f)
            for phenomenon_data in metrics_data:
                model_data.append(phenomenon_data[metric])
        metric_data.append(model_data)
    print(metric)
    plot_metric_heatmap_phenomena(np.array(metric_data).T, models, phenomena, save_path=f"../report/images/metric_heatmaps_phenomena/important_words/hypothesis_only_{metric}.pdf")
```
