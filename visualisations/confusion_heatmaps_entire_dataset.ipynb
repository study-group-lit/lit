```python
import json

import sys
sys.path.append('..')
from semplotlib import plot_confusion_heatmap
```

## default

```python
with open("../scripts/finetune_mnli/results/test.json", "r") as f:
    metrics_data = json.load(f)
    for task_data in metrics_data:
        print(task_data['task_name'])
        task_name = task_data['task_name'].replace("/", "_")
        plot_confusion_heatmap(task_data["confusion_matrix"], relative=True, save_path=f"../report/images/confusion_heatmaps/default/{task_name}.pdf")
```
