```python
from datasets import load_dataset
from transformers import pipeline
from evaluate import evaluator
import evaluate
```

```python
sick = load_dataset("sick")
```

```python
model = pipeline("text-classification", model="../models/roberta-base-mnli/")
```

```python
metrics = evaluate.combine(["accuracy", evaluate.load("f1", average="macro"), "matthews_correlation"])
```

```python
sick_validation =  sick["validation"].rename_column("sentence_A", "text").rename_column("sentence_B", "text_pair")
```

```python
results = evaluator("text-classification").compute(model, sick_validation, metric=metrics)

print(results)
```
